```python
def convert_to_integer(amount):
    unit_multipliers = {'Crore+': 10000000, 'Lac+': 100000, 'Thou+': 1000, 'Hund+': 100}

    for unit, multiplier in unit_multipliers.items():
        if unit in amount:
            numeric_part = float(amount.split(' ')[0])
            return int(numeric_part * multiplier)

    return int(amount)
data_train['Total Assets']=data_train['Total Assets'].apply(convert_to_integer)
data_train['Liabilities']=data_train['Liabilities'].apply(convert_to_integer)
data_test['Total Assets']=data_test['Total Assets'].apply(convert_to_integer)
data_test['Liabilities']=data_test['Liabilities'].apply(convert_to_integer)


```
